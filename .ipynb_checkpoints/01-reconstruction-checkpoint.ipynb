```python
import os

def extract_projection_number(filename):
    """
    Extracts the projection number from the filename.
    Assumes filenames are in the format 'XXXX_XXXX_XXXX_XXXX_XXXX_XXXX_XXXX.tif'
    where the projection number is the last segment before the file extension.
    """
    base_name = os.path.basename(filename)
    try:
        number_str = base_name.split('_')[-1].split('.')[0]
        return int(number_str)
    except (IndexError, ValueError):
        print(f"Failed to extract projection number from filename: {filename}")
        return None

def calculate_angle(projection_number, total_projections):
    """
    Calculates the angle for a given projection number.
    """
    return (360 / total_projections) * projection_number

def extract_angles_from_tiff_files(tiff_directory, total_projections):
    """
    Extracts angles for available TIFF files in the specified directory.
    """
    try:
        # List all files in the directory
        all_files = os.listdir(tiff_directory)
        print(f"Files in directory: {all_files}")

        # Filter for TIFF files
        tiff_files = [f for f in all_files if f.endswith('.tif')]
    except Exception as e:
        print(f"Error accessing directory {tiff_directory}: {e}")
        return [], []

    if not tiff_files:
        print(f"No TIFF files found in directory {tiff_directory}")
        return [], []

    # Extract projection numbers
    projection_numbers = []
    for f in tiff_files:
        proj_num = extract_projection_number(f)
        if proj_num is not None:
            projection_numbers.append(proj_num)

    if not projection_numbers:
        print("No valid projection numbers extracted from filenames.")
        return [], []

    # Calculate angles for each projection number
    angles = [calculate_angle(num, total_projections) for num in projection_numbers]

    return projection_numbers, angles

# Path to the directory containing the TIFF files
tiff_directory = "C:/Users/snn23kfl/abdulla data/tomo_2/"
total_projections = 1583

# Extract projection numbers and corresponding angles
projection_numbers, angles = extract_angles_from_tiff_files(tiff_directory, total_projections)

# Check if any angles were extracted
if projection_numbers and angles:
    # Print the projection numbers and their corresponding angles
    for proj_num, angle in zip(projection_numbers, angles):
        print(f"Projection: {proj_num}, Angle: {angle}")

    # Write angles to a file
    angles_file_path = os.path.join(tiff_directory, "angles.txt")
    with open(angles_file_path, 'w') as file:
        for angle in angles:
            file.write(f"{angle}\n")
else:
    print("No angles extracted.")

```
